# Building Blocks of Language

## **Ngữ cảnh (Context)**
Ngữ cảnh là cách các phần khác nhau trong ngôn ngữ kết hợp lại để truyền đạt một ý nghĩa cụ thể. Ngữ cảnh bao gồm các ***tham chiếu dài hạn(long-term references ), kiến thức về thế giới, và hiểu biết thông thường, cùng với nghĩa đen của từ và cụm từ.*** Ý nghĩa của một câu có thể thay đổi dựa trên ngữ cảnh, vì từ và cụm từ đôi khi có thể mang nhiều nghĩa khác nhau. Thông thường, ngữ cảnh được hình thành từ ngữ nghĩa học (semantics) và ngữ dụng học (pragmatics). Ngữ nghĩa học đề cập đến ý nghĩa trực tiếp của các từ và câu mà không cần đến ngữ cảnh bên ngoài. Ngữ dụng học bổ sung kiến thức về thế giới và ngữ cảnh bên ngoài của cuộc hội thoại để giúp chúng ta suy ra ý nghĩa ngụ ý. Các nhiệm vụ NLP phức tạp như phát hiện châm biếm, tóm tắt, và mô hình hóa chủ đề là một số nhiệm vụ sử dụng ngữ cảnh rất nhiều.

Long-term references (tham chiếu dài hạn) trong ngữ cảnh ngôn ngữ đề cập đến các thông tin hoặc khái niệm đã được nhắc đến hoặc hiểu trong một khoảng thời gian dài trước đó trong cuộc hội thoại, văn bản, hoặc thậm chí là trong nền tảng kiến thức chung của người giao tiếp.

## **Common knowledge(kiến thức chung)**
Một khía cạnh quan trọng của bất kỳ ngôn ngữ nào là "***kiến thức phổ thông***"—tập hợp các sự kiện mà hầu hết con người đều biết. Trong bất kỳ cuộc hội thoại nào, những sự kiện này được giả định là đã biết và vì thế không cần được đề cập rõ ràng, nhưng chúng vẫn ảnh hưởng đến ý nghĩa của câu.

Ví dụ, hãy xem xét hai câu sau:
1.   "Man bit dog."
2.   "Dog bit man."


Chúng ta đều hiểu rằng câu đầu tiên rất khó xảy ra, trong khi câu thứ hai là khả dĩ. Tại sao lại như vậy? Bởi vì chúng ta "biết" rằng con người rất ít khi cắn chó, trong khi chó lại thường có khả năng cắn người.

Kiến thức này không được nhắc đến trực tiếp trong câu, nhưng nó cần thiết để chúng ta có thể nhận định rằng câu đầu tiên là khó xảy ra, còn câu thứ hai là hợp lý. Con người sử dụng kiến thức phổ thông liên tục để hiểu và xử lý ngôn ngữ.

Trong ví dụ trên, hai câu có cấu trúc cú pháp rất giống nhau, nhưng máy tính lại gặp khó khăn trong việc phân biệt, vì nó thiếu đi kiến thức phổ thông mà con người có. Đây là một trong những thách thức chính trong NLP: làm thế nào để mã hóa tất cả những điều mà con người coi là kiến thức phổ thông vào một mô hình máy tính.

# **Machine Learning, Deep Learning, and NLP: An Overview**

## **Tiếp cận đến NLP**
Có nhiều cách khác nhau để xử lý bài toán NLP chung quy lại thì có 3 cách tiếp cận: heuristics(quy tắc), machine learning, and deep learning.


## **Heuristics-Based NLP**
Tương tự như các hệ thống AI ban đầu, những nỗ lực đầu tiên trong việc thiết kế các hệ thống NLP dựa trên việc xây dựng các quy tắc để thực hiện nhiệm vụ. Điều này yêu cầu các nhà phát triển phải có chuyên môn trong lĩnh vực để xây dựng các quy tắc có thể được tích hợp vào chương trình. Các hệ thống như vậy cũng cần các tài nguyên như từ điển và từ đồng nghĩa, thường được biên soạn và số hóa qua một thời gian dài. Một ví dụ về việc thiết kế quy tắc để giải quyết vấn đề NLP bằng cách sử dụng các tài nguyên này là phân tích cảm xúc dựa trên từ vựng (lexicon-based sentiment analysis), trong đó sử dụng số lượng từ tích cực và tiêu cực trong văn bản để suy ra cảm xúc của văn bản. Chúng ta sẽ đề cập đến điều này trong Chương 4.

Bên cạnh từ điển và từ đồng nghĩa, các cơ sở tri thức phức tạp hơn đã được xây dựng để hỗ trợ NLP nói chung và NLP dựa trên quy tắc nói riêng. Một ví dụ là ***WordNet***, một cơ sở dữ liệu về từ vựng và mối quan hệ ngữ nghĩa giữa chúng. Một số mối quan hệ tiêu biểu bao gồm:


1.   Synonyms (từ đồng nghĩa): Các từ có ý nghĩa tương tự nhau.
2.   Hyponyms (từ loại con): Nắm bắt mối quan hệ kiểu "là một loại". Ví dụ: bóng chày, sumo, và tennis đều là các hyponyms của thể thao.
3.   Meronyms (từ bộ phận): Nắm bắt mối quan hệ "là một phần của". Ví dụ: tay và chân là meronyms của cơ thể.
Tất cả những thông tin này trở nên hữu ích khi xây dựng các hệ thống dựa trên quy tắc về ngôn ngữ.




***Biểu thức chính quy (Regular expressions - regex)*** là một công cụ tuyệt vời trong phân tích văn bản và xây dựng các hệ thống dựa trên quy tắc.

Regex là tập hợp các ký tự hoặc một mẫu được sử dụng để khớp và tìm các chuỗi con trong văn bản.

Ví dụ: Regex sau đây
^([a-zA-Z0-9_\-\.]+)@([a-zA-Z0-9_\-\.]+)\.([a-zA-Z]{2,5})$
được dùng để tìm tất cả các địa chỉ email trong một đoạn văn bản.

Regex là một cách tuyệt vời để tích hợp kiến thức miền (domain knowledge) vào hệ thống NLP. Ví dụ, nếu chúng ta muốn xây dựng một hệ thống tự động xác định sản phẩm trong các khiếu nại của khách hàng, có thể sử dụng regex để khớp các mã sản phẩm với các tên thương hiệu liên quan.

Regex cũng được sử dụng rộng rãi trong phần mềm NLP như StanfordCoreNLP, với TokensRegex là một khung công cụ định nghĩa các biểu thức chính quy.

***Ngữ pháp phi ngữ cảnh (CFG)***

Ngữ pháp phi ngữ cảnh (Context-free grammar - CFG) là một loại ngữ pháp hình thức được sử dụng để mô hình hóa ngôn ngữ tự nhiên. Nó được phát minh bởi Giáo sư Noam Chomsky. CFG có thể nắm bắt các thông tin phức tạp và có tính phân cấp mà regex không thể làm được.

JAPE (Java Annotation Patterns Engine) là một ngôn ngữ ngữ pháp hỗ trợ cả regex và CFG, được sử dụng trong hệ thống NLP dựa trên quy tắc như GATE (General Architecture for Text Engineering). Ví dụ, JAPE và GATE đã được sử dụng để trích xuất thông tin về các quy trình cấy ghép máy tạo nhịp từ các báo cáo y tế.

Quy tắc và heuristic trong NLP hiện đại
Quy tắc và heuristic vẫn đóng vai trò quan trọng trong toàn bộ vòng đời của các dự án NLP. Ở giai đoạn đầu, chúng là một cách hiệu quả để xây dựng phiên bản đầu tiên của hệ thống NLP, giúp bạn hiểu rõ hơn về vấn đề cần giải quyết.

Ở giai đoạn sản xuất, quy tắc và heuristic giúp lấp đầy các lỗ hổng trong hệ thống. Ví dụ, một hệ thống y tế sử dụng NLP để phân tích hồ sơ bệnh án có thể đưa ra quyết định sai lầm nghiêm trọng nếu không đề xuất một xét nghiệm quan trọng, điều này có thể ảnh hưởng đến tính mạng. Quy tắc và heuristic là cách hữu ích để ngăn chặn các sai lầm như vậy.

Tiếp theo, chúng ta sẽ chuyển sang các kỹ thuật học máy được sử dụng trong NLP.

## **Machine Learning for NLP**
Các kỹ thuật học máy được áp dụng lên dữ liệu văn bản tương tự như cách chúng được áp dụng lên các dạng dữ liệu khác, chẳng hạn như hình ảnh, âm thanh, hoặc dữ liệu có cấu trúc. Các phương pháp học máy có giám sát như classification (phân loại) và regression (hồi quy)  được sử dụng rộng rãi trong nhiều nhiệm vụ NLP.

Ví dụ:

1.   Nhiệm vụ phân loại trong NLP có thể là phân loại bài báo thành các chủ đề như thể thao hoặc chính trị.
2.   Kỹ thuật hồi quy, cung cấp dự đoán dưới dạng số, có thể được sử dụng để ước tính giá cổ phiếu dựa trên các thảo luận trên mạng xã hội về cổ phiếu đó.
3.   Các thuật toán không giám sát, như clustering (phân cụm), có thể được sử dụng để nhóm các tài liệu văn bản lại với nhau.

Mọi phương pháp học máy trong NLP, dù có giám sát hay không, đều bao gồm ba bước chính:


1.   Trích xuất đặc trưng từ văn bản.
2.   Sử dụng biểu diễn đặc trưng để huấn luyện một mô hình.
3.   Đánh giá và cải thiện mô hình.

Chúng ta sẽ tìm hiểu chi tiết về các biểu diễn đặc trưng của văn bản trong **Chương 3** và cách đánh giá trong **Chương 2**.

***Một số thuật toán học máy phổ biến trong NLP***
### **Naive Bayes**
Đây là một thuật toán kinh điển cho các bài toán phân loại, dựa trên định lý Bayes.

Nó tính xác suất xảy ra của một nhãn (class) dựa trên tập các đặc trưng của dữ liệu đầu vào. Thuật toán này giả định rằng mỗi đặc trưng là độc lập với các đặc trưng khác.

Ví dụ: Trong bài toán phân loại tin tức, văn bản có thể được biểu diễn bằng số lần xuất hiện của các từ thuộc lĩnh vực thể thao hoặc chính trị. Naive Bayes giả định rằng các từ này không có mối tương quan với nhau.

Mặc dù giả định này có thể không chính xác trong nhiều trường hợp, Naive Bayes vẫn thường được sử dụng như một thuật toán khởi đầu vì tính đơn giản và tốc độ.


### **Support vector machine**
SVM là một thuật toán phân loại phổ biến, tập trung vào việc học đường biên quyết định (decision boundary) để phân tách các lớp khác nhau.

Đường biên này có thể là tuyến tính hoặc phi tuyến. Ví dụ: Một đường biên phi tuyến có thể là hình tròn.

SVM tối ưu hóa đường biên sao cho khoảng cách giữa các điểm thuộc các lớp khác nhau là lớn nhất, giúp nó có khả năng xử lý tốt dữ liệu nhiễu và biến đổi.

Điểm mạnh: Độ chính xác cao và khả năng chống nhiễu.
Điểm yếu: Thời gian huấn luyện lâu và không phù hợp với dữ liệu lớn.


### **Hidden Markov Model**
HMM là một mô hình thống kê giả định rằng có một quá trình ẩn sinh ra dữ liệu mà chúng ta chỉ có thể quan sát.

Ví dụ: Trong bài toán gắn thẻ từ loại (POS tagging), các từ trong câu được giả định tuân theo một ngữ pháp ẩn (grammar). Các trạng thái ẩn chính là các từ loại (như danh từ, động từ, tính từ), và chúng ta chỉ quan sát được câu văn.

HMM còn giả định tính chất Markov, nghĩa là mỗi trạng thái ẩn phụ thuộc vào trạng thái trước đó. Do ngôn ngữ có tính chất tuần tự, HMM rất mạnh mẽ trong việc mô hình hóa dữ liệu văn bản.


### **Conditional random fields**
CRF là một thuật toán được sử dụng cho dữ liệu tuần tự.

Ví dụ: Trong bài toán gắn thẻ từ loại, CRF gắn thẻ từng từ dựa trên ngữ cảnh của các từ xung quanh. CRF mạnh hơn HMM vì nó không chỉ xem xét ngữ cảnh tuần tự mà còn kết hợp nhiều yếu tố khác trong mô hình.
CRF thường được sử dụng trong các tác vụ đòi hỏi xử lý chuỗi dữ liệu liên tiếp, và chúng thường vượt trội so với HMM trong các bài toán như POS tagging.

## **Deep Learning for NLP**
 Ngôn ngữ tự nhiên vốn dĩ rất phức tạp và không có cấu trúc, vì vậy chúng ta cần các mô hình với khả năng biểu diễn và học tập tốt hơn để hiểu và giải quyết các nhiệm vụ NLP.

### **Recurrent neural networks**
Ngôn ngữ tự nhiên có tính chất tuần tự. Ví dụ, một câu trong ngôn ngữ (như tiếng Anh) được đọc từ trái sang phải. RNNs được thiết kế đặc biệt để xử lý và học từ dữ liệu tuần tự.

RNN có các đơn vị nơ-ron có khả năng ghi nhớ những gì đã xử lý trước đó.
Bộ nhớ này là tạm thời và được cập nhật sau mỗi bước thời gian khi RNN đọc các từ tiếp theo.
Ứng dụng:

Phân loại văn bản, nhận diện thực thể (named entity recognition), dịch máy, và tạo văn bản (text generation).

### **Long short-term memory**
Để khắc phục hạn chế về khả năng ghi nhớ dài hạn của RNN, LSTM đã được phát triển.

LSTM loại bỏ ngữ cảnh không liên quan và chỉ ghi nhớ những phần ngữ cảnh cần thiết để giải quyết nhiệm vụ.

Điều này giảm tải việc phải lưu trữ toàn bộ ngữ cảnh dài hạn trong một biểu diễn duy nhất.

LSTM đã thay thế RNN trong hầu hết các ứng dụng NLP do khả năng xử lý ngữ cảnh dài tốt hơn. Một biến thể khác là GRU (Gated Recurrent Units), thường được sử dụng trong các tác vụ tạo văn bản.

### **Convolutional neural networks**
CNNs được sử dụng phổ biến trong thị giác máy tính (như phân loại hình ảnh), nhưng chúng cũng thành công trong các nhiệm vụ NLP, đặc biệt là phân loại văn bản.

Văn bản được biểu diễn dưới dạng ma trận (n × d), trong đó:
n là số từ trong câu.
d là kích thước của vector từ (word vector).
Ma trận này có thể được xử lý tương tự như hình ảnh trong CNN.
Ưu điểm:

CNN sử dụng cửa sổ ngữ cảnh để xem xét các nhóm từ liền kề, giúp phân tích các cụm từ và ngữ nghĩa tổng thể tốt hơn.

### **Transformers**
Transformers là kiến trúc tiên tiến nhất trong NLP, đạt được **state of the art** trong hầu hết các nhiệm vụ NLP trong vài năm qua.

Thay vì xử lý tuần tự như RNN, transformers sử dụng self-attention để xem xét toàn bộ ngữ cảnh xung quanh một từ.

Ví dụ: Từ "bank" có nghĩa khác nhau trong hai ngữ cảnh "finance" (ngân hàng) và "river" (bờ sông).

Transformers đã dẫn đến sự phát triển của các mô hình như **BERT** (Bidirectional Encoder Representations from Transformers).

BERT được huấn luyện trước trên lượng dữ liệu khổng lồ (hơn 40GB) và sau đó được tinh chỉnh (fine-tuning) để giải quyết các nhiệm vụ cụ thể như phân loại văn bản, trả lời câu hỏi, và trích xuất thực thể.1-16.png

### **Autoencoders**
Autoencoders là một loại mạng nơ-ron được sử dụng để học biểu diễn vector nén của đầu vào.

Ví dụ: Nếu muốn biểu diễn một đoạn văn bản bằng một vector, autoencoder có thể học ánh xạ từ văn bản sang vector này.

Quá trình này không cần nhãn dữ liệu, vì vậy đây là một dạng học không giám sát.
Biến thể của autoencoders, như LSTM autoencoders, xử lý tốt hơn dữ liệu tuần tự như văn bản.

## **Tại sao học sâu không phải luôn là giải pháp lý tưởng cho NLP?**

Mặc dù học sâu đạt được nhiều thành tựu trong NLP, nhưng trong nhiều trường hợp thực tế, nó không phải là lựa chọn tốt nhất do:

Yêu cầu tài nguyên tính toán lớn.
Dữ liệu cần thiết để huấn luyện rất nhiều.
Khó triển khai và bảo trì trong môi trường sản xuất
